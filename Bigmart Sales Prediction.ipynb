```python
# Replacing na values in 'Item_Weight' with mean

df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].mean())
```

```python
# Replacing na values in 'Outlet_Size' with mode

df['Outlet_Size'] = df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])
```

## Outlier treatment

```python
# Returns Outliers

def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers, lower_bound, upper_bound
```

```python
# Print number of Outliers for each numeric columns

col_num = df.select_dtypes(include='float64')

for column in col_num :
    print(column, '\n', find_outliers(df, column)[0].count())
```

```python
# Replacing outlier values with calculated upper and lower threshold values respectively

upper_thresh = find_outliers(df, 'Item_Outlet_Sales')[2]
lower_thresh = find_outliers(df, 'Item_Outlet_Sales')[1]

df['Item_Outlet_Sales'] = np.where(df['Item_Outlet_Sales'] > upper_thresh, upper_thresh, df['Item_Outlet_Sales'])
df['Item_Outlet_Sales'] = np.where(df['Item_Outlet_Sales'] < lower_thresh, lower_thresh, df['Item_Outlet_Sales'])
```

```python
print(upper_thresh)
print(lower_thresh)
```

## Splitting into train/test set

```python
# Defining Independent and Dependent variables

X = df.drop(['Item_Identifier', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Item_Outlet_Sales'], axis= 1)
y = df['Item_Outlet_Sales']
```

```python
# Label Encoding categorical variables into integers

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

for column in columns :
    X[column] = le.fit_transform(X[column])
```

```python
# Splitting into Train and Test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
X_train = pd.get_dummies(X_train, drop_first=True, dtype='int8')
X_test = pd.get_dummies(X_test, drop_first=True, dtype='int8')
```

# Linear Regression

```python
# Appying Linear Regression Model onto Train set

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
y_pred
```

```python
# Calculating R2 Score

from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

# Polynomial Regression

```python
# Calculating Polynomial Variables for train set

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X_train)

# Appying Polynomial Regression Model onto Train set

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y_train)
```

```python
y_pred = poly_reg.predict(poly.transform(X_test))
y_pred
```
