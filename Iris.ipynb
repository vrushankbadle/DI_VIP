```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataset as df

df = pd.read_csv('IRIS.csv')
df.head()
```

```python
df.info()
```

# Splitting into train test

```python
# Defining Dependent and Independent variables

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Splitting into Train and Test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Applying Scaling on independent variables

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Applying Logistic Regression Model onto Train set

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Printing Predicted values and Observed values from Test set

y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(-1,1), y_test.reshape(-1,1)), axis=1 ))
```

```python
# Evaluating Model by Confusion Matrix and Accuracy Score

from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

Model gives 97% accuracy and hence is a very good model
